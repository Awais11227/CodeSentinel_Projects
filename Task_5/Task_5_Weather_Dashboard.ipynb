```python
import tkinter as tk
from tkinter import ttk
from datetime import datetime

demo_data = {
    "name": "London",
    "sys": {"country": "GB", "sunrise": 1693116000, "sunset": 1693168800},
    "main": {"temp": 18, "humidity": 65},
    "weather": [{"description": "light rain"}],
    "wind": {"speed": 4.5},
    "clouds": {"all": 78},
    "visibility": 9000
}

# 5-Day forecast
forecast_data = [
    {"day": "Mon", "date": "Aug 27", "temp": "19¬∞C", "condition": "‚òÅÔ∏è Cloudy"},
    {"day": "Tue", "date": "Aug 28", "temp": "20¬∞C", "condition": "üåßÔ∏è Rainy"},
    {"day": "Wed", "date": "Aug 29", "temp": "21¬∞C", "condition": "‚òÄÔ∏è Sunny"},
    {"day": "Thu", "date": "Aug 30", "temp": "17¬∞C", "condition": "‚õÖ Partly Cloudy"},
    {"day": "Fri", "date": "Aug 31", "temp": "16¬∞C", "condition": "üå¶Ô∏è Light Rain"},
]

# ==================== Main Window ====================
root = tk.Tk()
root.title("Weather Dashboard Pro")
root.geometry("900x600")  
root.configure(bg="#2E4A62")

title_label = tk.Label(root, text="Weather Dashboard", font=("Arial", 20, "bold"), bg="#2E4A62", fg="white")
title_label.pack(pady=10)

subtitle_label = tk.Label(root, text="Prepared by Awais Manzoor", font=("Arial", 11, "bold"), bg="#2E4A62", fg="orange")
subtitle_label.pack()

demo_label = tk.Label(root, text="Demo Mode ‚Ä¢ No Internet Required", font=("Arial", 9), bg="#2E4A62", fg="lightgray")
demo_label.pack(pady=(0, 10))

# ==================== Button Frame ====================
btn_frame = tk.Frame(root, bg="#2E4A62")
btn_frame.pack(pady=5)

def load_weather():
    labels["Location:"].config(text=f"üìç {demo_data['name']}, {demo_data['sys']['country']}")
    labels["Temperature:"].config(text=f"üå°Ô∏è {demo_data['main']['temp']}¬∞C")
    labels["Condition:"].config(text=f"üå§Ô∏è {demo_data['weather'][0]['description'].title()}")
    labels["Humidity:"].config(text=f"üíß {demo_data['main']['humidity']}%")
    labels["Wind Speed:"].config(text=f"üí® {demo_data['wind']['speed']} m/s")
    labels["Cloudiness:"].config(text=f"‚òÅÔ∏è {demo_data['clouds']['all']}%")
    labels["Visibility:"].config(text=f"üëÅÔ∏è {demo_data['visibility'] / 1000:.1f} km")
    sunrise = datetime.fromtimestamp(demo_data["sys"]["sunrise"]).strftime("%I:%M %p")
    sunset = datetime.fromtimestamp(demo_data["sys"]["sunset"]).strftime("%I:%M %p")
    labels["Sunrise:"].config(text=f"üåÖ {sunrise}")
    labels["Sunset:"].config(text=f"üåá {sunset}")

    
    for i, row in enumerate(forecast_rows):
        row["day"].config(text=forecast_data[i]["day"])
        row["date"].config(text=forecast_data[i]["date"])
        row["temp"].config(text=forecast_data[i]["temp"])
        row["condition"].config(text=forecast_data[i]["condition"])


load_button = tk.Button(
    btn_frame,
    text="üå§ Load Demo Weather",
    font=("Arial", 12, "bold"),
    bg="#4CAF50",
    fg="white",
    activebackground="#388E3C",
    activeforeground="white",
    relief="flat",
    width=18,
    height=2,
    command=load_weather,
    cursor="hand2"
)
load_button.grid(row=0, column=0, padx=10)

exit_button = tk.Button(
    btn_frame,
    text=" Exit",
    font=("Arial", 12, "bold"),
    bg="#E74C3C",
    fg="white",
    activebackground="#C0392B",
    activeforeground="white",
    relief="flat",
    width=10,
    height=2,
    command=root.destroy,
    cursor="hand2"
)
exit_button.grid(row=0, column=1, padx=10)

# Hover effects
def on_enter(btn, color):
    btn.config(bg=color)

def on_leave(btn, color):
    btn.config(bg=color)

load_button.bind("<Enter>", lambda e: on_enter(load_button, "#388E3C"))
load_button.bind("<Leave>", lambda e: on_leave(load_button, "#4CAF50"))

exit_button.bind("<Enter>", lambda e: on_enter(exit_button, "#C0392B"))
exit_button.bind("<Leave>", lambda e: on_leave(exit_button, "#E74C3C"))

main_frame = tk.Frame(root, bg="#2E4A62")
main_frame.pack(padx=20, pady=10, fill="both", expand=True)


frame_weather = ttk.LabelFrame(main_frame, text=" üå§ Current Weather ", padding=15)
frame_weather.pack(side="left", padx=(0, 10), fill="y")

field_colors = {
    "Location:": "white",
    "Temperature:": "orange",
    "Condition:": "lightgreen",
    "Humidity:": "cyan",
    "Wind Speed:": "lightblue",
    "Cloudiness:": "violet",
    "Visibility:": "yellow",
    "Sunrise:": "gold",
    "Sunset:": "red"
}

labels = {}
for field, color in field_colors.items():
    lbl = tk.Label(
        frame_weather,
        text=f"{field} ---",
        font=("Arial", 11, "bold"),
        fg=color,
        bg="#2E4A62",
        justify="left"
    )
    lbl.pack(anchor="w", pady=4)
    labels[field] = lbl

frame_forecast = ttk.LabelFrame(main_frame, text=" üìÖ 5-Day Forecast ", padding=15)
frame_forecast.pack(side="right", padx=(10, 0), fill="both", expand=True)

# Header
header_frame = tk.Frame(frame_forecast, bg="#34495E")
header_frame.pack(fill="x", pady=(0, 5))

tk.Label(header_frame, text="Day", font=("Arial", 10, "bold"), bg="#34495E", fg="cyan", width=8).pack(side="left", padx=5)
tk.Label(header_frame, text="Date", font=("Arial", 10, "bold"), bg="#34495E", fg="yellow", width=8).pack(side="left", padx=5)
tk.Label(header_frame, text="Temp", font=("Arial", 10, "bold"), bg="#34495E", fg="orange", width=8).pack(side="left", padx=5)
tk.Label(header_frame, text="Condition", font=("Arial", 10, "bold"), bg="#34495E", fg="lightgreen", width=12).pack(side="left")

# Rows
forecast_rows = []
for i in range(5):
    row_frame = tk.Frame(frame_forecast, bg="#2E4A62")
    row_frame.pack(fill="x", pady=2)

    day_lbl = tk.Label(row_frame, text="---", font=("Arial", 10), bg="#2E4A62", fg="white", width=8)
    day_lbl.pack(side="left", padx=5)

    date_lbl = tk.Label(row_frame, text="---", font=("Arial", 10), bg="#2E4A62", fg="lightgray", width=8)
    date_lbl.pack(side="left", padx=5)

    temp_lbl = tk.Label(row_frame, text="---", font=("Arial", 10), bg="#2E4A62", fg="orange", width=8)
    temp_lbl.pack(side="left", padx=5)

    cond_lbl = tk.Label(row_frame, text="---", font=("Arial", 10), bg="#2E4A62", fg="lightgreen", width=12)
    cond_lbl.pack(side="left")

    
    forecast_rows.append({
        "day": day_lbl,
        "date": date_lbl,
        "temp": temp_lbl,
        "condition": cond_lbl
    })


footer = tk.Label(root, text="", bg="#2E4A62")
footer.pack(pady=10)

# Run the app
root.mainloop()
```
